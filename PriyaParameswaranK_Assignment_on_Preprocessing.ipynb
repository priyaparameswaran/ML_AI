Null values are present in Age, Cabin and Embarked fields

Out of 891 rows, 687 missing values are seen in Cabin field. It would be difficult to handle the null values in field, as it may show variation in the model output, if mode value is used. So, this can be dropped.

```python
df1=df.drop(['Cabin'],axis=1)
df1
```

A constant value 0 cannot be added to Age field, so we are not using simple imouter function to handle Age value

```python
df1['Age']=df1['Age'].fillna(df1['Age'].median())
```

```python
df1['Embarked']=df1['Embarked'].fillna(df1['Embarked'].mode().iloc[0])
df1.isna().sum()
```

Thus, the missing values are handled


```python
df1.dtypes
```

As we need to male the feature values under int/float datatype, we need to do encoding to categorical fields.

We have 3 object fields. As the Name field will not have much significance in the model, this can be dropped.
The other fields are Sex and Embarked. Now, we can check for the number of unique values in each field

```python
df1.nunique()
```

```python
#Label encoding for the fields "Sex" and "Embarked"
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df1['Sex']= label_encoder.fit_transform(df1['Sex'])
df1['Sex'].unique()
```

```python
df1['Embarked']= label_encoder.fit_transform(df1['Embarked'])
df1['Embarked'].unique()
```
